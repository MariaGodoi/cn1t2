1) A densidade do gás nitrogênio foi medida a 300 K, constante, em um intervalo de pressões. A partir destes dados, calcule a massa atômica do átomo de nitrogênio. Os dados se encontram em uma planilha em anexo.

```
Pressão (atm),Densidade (kg/m3)
"0,01","0,01138"
"0,51","0,58041"
"1,01","1,1495"
"1,51","1,7188"
"2,01","2,2881"
"2,51","2,8576"
"3,01","3,4271"
```


````python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
pressao = np.array([0.01, 0.51, 1.01, 1.51, 2.01, 2.51, 3.01])
densidade = np.array([0.01138, 0.58041, 1.1495, 1.7188, 2.2881, 2.8576, 3.4271])

# Ajuste linear manual (y = mx + b)
n = len(pressao)
sum_x = np.sum(pressao)
sum_y = np.sum(densidade)
sum_xy = np.sum(pressao * densidade)
sum_x2 = np.sum(pressao**2)

slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
intercept = (sum_y - slope * sum_x) / n

# Cálculos
T = 300
R = 0.082057
M = slope * R * T
massa_atomica = M / 2

#print(f"Slope: {slope:.6f}")#
print(f"Massa atômica do N: {massa_atomica:.4f} g/mol")

# Gráfico simples
plt.plot(pressao, densidade, 'bo', label='Dados')
plt.plot(pressao, slope*pressao + intercept, 'r-', label='Ajuste')
plt.xlabel('Pressão (atm)')
plt.ylabel('Densidade (g/L)')
plt.legend()
plt.grid(True)
plt.show()
````
