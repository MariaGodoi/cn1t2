1) A densidade do gás nitrogênio foi medida a 300 K, constante, em um intervalo de pressões. A partir destes dados, calcule a massa atômica do átomo de nitrogênio. Os dados se encontram em uma planilha em anexo.

1000179927.jpg


```python
# Análise do Gás Nitrogênio
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Dados experimentais
pressao = np.array([0.01, 0.51, 1.01, 1.51, 2.01, 2.51, 3.01])  # atm
densidade = np.array([0.01138, 0.58041, 1.1495, 1.7188, 2.2881, 2.8576, 3.4271])  # g/L

# Ajuste linear
coef_angular, intercepto, r_value, p_value, std_err = stats.linregress(pressao, densidade)

# Cálculo da massa atômica
R = 0.082057  # L·atm·mol⁻¹·K⁻¹
T = 300       # K
M = coef_angular * R * T
massa_atomica = M / 2

# Resultados
print("📊 RESULTADOS DA ANÁLISE")
print("=" * 40)
print(f"Coeficiente angular: {coef_angular:.6f} g·L⁻¹·atm⁻¹")
print(f"Coeficiente linear: {intercepto:.6f} g·L⁻¹")
print(f"Coeficiente R²: {r_value**2:.6f}")
print(f"Massa molar do N₂: {M:.4f} g·mol⁻¹")
print(f"Massa atômica do nitrogênio: {massa_atomica:.4f} g·mol⁻¹")
print("=" * 40)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(pressao, densidade, color='blue', s=80, label='Dados experimentais')
plt.plot(pressao, coef_angular * pressao + intercepto,
         color='red', linewidth=2,
         label=f'ρ = {coef_angular:.4f}P + {intercepto:.4f}')

plt.xlabel('Pressão (atm)', fontsize=12, fontweight='bold')
plt.ylabel('Densidade (g/L)', fontsize=12, fontweight='bold')
plt.title('Densidade do Nitrogênio vs Pressão (300 K)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend()
plt.tight_layout()

plt.show()

# Tabela comparativa
print("\n📋 COMPARAÇÃO: Experimental vs Calculado")
print("Pressão (atm) | Experimental (g/L) | Calculado (g/L)")
print("-" * 55)
for i in range(len(pressao)):
    calc = coef_angular * pressao[i] + intercepto
    print(f"{pressao[i]:10.2f}     | {densidade[i]:15.5f}     | {calc:15.5f}")
```

```python
# Análise do Gás Nitrogênio (método com extrapolação)
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
pressao = np.array([0.01, 0.51, 1.01, 1.51, 2.01, 2.51, 3.01])  # atm
densidade = np.array([0.01138, 0.58041, 1.1495, 1.7188, 2.2881, 2.8576, 3.4271])  # g/L

# Cálculo de ρ/P
rho_sobre_p = densidade / pressao

# Regressão linear: rho/P vs pressão
coef_angular, coef_linear = np.polyfit(pressao, rho_sobre_p, 1)
# A reta é: (ρ/P) = coef_angular * P + coef_linear

# Constantes
R = 0.082057  # L·atm·mol⁻¹·K⁻¹
T = 300       # K

# Extrair massa molar via extrapolação para P→0
media_extrapolada = coef_linear  # intercepto da regressão
M = media_extrapolada * R * T    # g/mol (mol od N₂)
massa_atomica = M / 2            # g/mol (átomo de N)

# Resultados
print("📊 RESULTADOS (método com extrapolação)")
print("=" * 50)
print(f"Coeficiente angular (ρ/P vs P): {coef_angular:.8f} g·L⁻¹·atm⁻²")
print(f"Coeficiente linear (P→0): {coef_linear:.6f} g·L⁻¹·atm⁻¹")
print(f"Massa molar do N₂ (via extrapolação): {M:.4f} g·mol⁻¹")
print(f"Massa atômica do nitrogênio: {massa_atomica:.4f} g·mol⁻¹")
print("=" * 50)

# Gráfico com extrapolação
plt.figure(figsize=(10, 6))
plt.scatter(pressao, rho_sobre_p, color='blue', s=80, label='Dados experimentais (ρ/P)')
x_fit = np.linspace(0, max(pressao), 100)
y_fit = coef_angular * x_fit + coef_linear
plt.plot(x_fit, y_fit, color='red', linewidth=2, label='Regressão linear (ρ/P vs P)')
plt.axhline(coef_linear, color='green', linestyle='--', label=f'Extrapolação P→0: {coef_linear:.4f}')

plt.xlabel('Pressão (atm)', fontsize=12, fontweight='bold')
plt.ylabel('ρ/P (g·L⁻¹·atm⁻¹)', fontsize=12, fontweight='bold')
plt.title('Extrapolação de ρ/P para P → 0 (300 K)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend()
plt.tight_layout()
plt.show()

# Tabela comparativa
print("\n📋 TABELA: Experimental vs Regressão")
print("Pressão (atm) | ρ/P experimental | ρ/P pela regressão")
print("-" * 55)
for P, val in zip(pressao, rho_sobre_p):
    val_reg = coef_angular * P + coef_linear
    print(f"{P:12.2f}     | {val:15.5f}     | {val_reg:15.5f}")
```
