```python

# QUESTÃO 8 – DIAGRAMA DE FASE MgO–NiO

import numpy as np
import matplotlib.pyplot as plt

# Dados da tabela fornecida
T = np.array([1960, 2200, 2400, 2600, 2800])
x = np.array([0, 0.35, 0.60, 0.83, 1])   # fração sólida
y = np.array([0, 0.18, 0.38, 0.65, 1])   # fração líquida

# -----------------------------------------------
# a) Ponto de fusão para x = 0.30
# -----------------------------------------------
x0 = 0.30
T_x30 = np.interp(x0, x, T)

# -----------------------------------------------
# b) Composição das fases e proporções a 2200 ºC
# -----------------------------------------------
T_target = 2200

x_s = np.interp(T_target, T, x)  # sólido na temperatura
y_l = np.interp(T_target, T, y)  # líquido na temperatura

# Frações em cada fase (regra da alavanca)
f_l = (x0 - x_s) / (y_l - x_s)
f_s = 1 - f_l

# -----------------------------------------------
# c) Temp. onde líquido y = 0.70 começa a solidificar
# -----------------------------------------------
y0 = 0.70
T_y70 = np.interp(y0, y, T)

# -----------------------------------------------
# Impressão dos resultados
# -----------------------------------------------

print("========== RESULTADOS ==========\n")

print(f"a) Ponto de fusão para x=0.30:   T = {T_x30:.2f} °C")
print("\n---------------------------------\n")

print("b) A 2200 °C, para x=0.30:")
print(f"   - Composição sólido: x_s = {x_s:.3f}")
print(f"   - Composição líquido: y_l = {y_l:.3f}")
print(f"   - Fração sólida  = {f_s:.4f}  (ou {f_s*100:.1f}%) ")
print(f"   - Fração líquida = {f_l:.4f}  (ou {f_l*100:.1f}%) ")
print("\n---------------------------------\n")

print(f"c) Temperatura onde y=0.70 começa a solidificar: T = {T_y70:.2f} °C")

# -----------------------------------------------
# Gráfico do diagrama T–x
# -----------------------------------------------

plt.figure(figsize=(8,6))
plt.plot(x, T, 'o-', label='Sólido (fração sólida x)')
plt.plot(y, T, 's-', label='Líquido (fração líquida y)')

plt.axvline(0.30, linestyle='--', color='gray', alpha=0.5)
plt.axhline(T_x30, linestyle='--', color='gray', alpha=0.5)

plt.xlabel("Fração molar MgO")
plt.ylabel("Temperatura (°C)")
plt.title("Diagrama de Fases MgO–NiO")
plt.legend()
plt.grid(True)
plt.show()
```
