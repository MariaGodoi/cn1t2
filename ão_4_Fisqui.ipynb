```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.optimize import curve_fit

# Dados fornecidos
T = np.array([13.74, 16.25, 20.43, 27.73, 41.11, 52.72, 68.15, 82.96, 95.71])  # Temperatura (K)
Cp = np.array([0.171, 0.286, 0.626, 1.615, 4.346, 7.032, 10.48, 13.28, 15.33])  # Cp (cal K)

# Calculando Cp/T
Cp_T = Cp / T

# a) Ajuste de polinmio de terceira ordem que passa pela origem
def poly_3rd_order(x, a, b, c):
    """Polinmio de terceira ordem que passa pela origem: f(x) = a*x + b*x + c*x"""
    return a*x + b*x**2 + c*x**3

# Ajuste dos dados
params, covariance = curve_fit(poly_3rd_order, T, Cp_T)
a, b, c = params

print("Parmetros do polinmio de terceira ordem:")
print(f"a = {a:.6f}, b = {b:.6f}, c = {c:.6f}")

# Funo ajustada
def fitted_function(x):
    return poly_3rd_order(x, a, b, c)

# Clculo da entropia a 90 K (rea sob a curva de 0 a 90 K)
S_90K, error = integrate.quad(fitted_function, 0, 90)
print(f"\na) Entropia do NaSO a 90 K: {S_90K:.4f} cal K")

# b) Ajuste da funo aT em baixas temperaturas
def cubic_law(T, a):
    """Lei cbica: Cp/T = a*T"""
    return a * T**3

# Vamos testar o ajuste em diferentes intervalos de temperatura
temperatures_to_test = [
    (0, 20), (0, 25), (0, 30), (0, 35), (0, 40),
    (0, 45), (0, 50), (0, 55), (0, 60)
]

print("\nb) Investigao da validade da aproximao aT:")
print("Intervalo\tParmetro a\t\tR")

best_interval = None
best_r2 = -np.inf

for i, (T_min, T_max) in enumerate(temperatures_to_test):
    # Selecionar dados no intervalo
    mask = (T >= T_min) & (T <= T_max)
    T_subset = T[mask]
    Cp_T_subset = Cp_T[mask]

    if len(T_subset) < 2:
        continue

    # Ajuste da funo cbica
    try:
        a_param, _ = curve_fit(cubic_law, T_subset, Cp_T_subset)

        # Calcular R
        y_pred = cubic_law(T_subset, a_param[0])
        ss_res = np.sum((Cp_T_subset - y_pred)**2)
        ss_tot = np.sum((Cp_T_subset - np.mean(Cp_T_subset))**2)
        r2 = 1 - (ss_res / ss_tot) if ss_tot != 0 else 0

        print(f"{T_min}-{T_max} K\t{a_param[0]:.6e}\t\t{r2:.4f}")

        # Encontrar o melhor intervalo (maior R)
        if r2 > best_r2:
            best_r2 = r2
            best_interval = (T_min, T_max)
            best_a = a_param[0]

    except:
        continue

print(f"\nMelhor intervalo para aproximao aT: {best_interval[0]}-{best_interval[1]} K")
print(f"Parmetro a timo: {best_a:.6e}")
print(f"Coeficiente de determinao R: {best_r2:.4f}")

# Visualizao dos resultados
plt.figure(figsize=(12, 8))

# Grfico 1: Dados experimentais e ajuste polinomial
plt.subplot(2, 2, 1)
T_plot = np.linspace(0, 100, 500)
plt.plot(T, Cp_T, 'bo', label='Dados experimentais')
plt.plot(T_plot, fitted_function(T_plot), 'r-', label=f'Ajuste polinomial: {a:.3e}T + {b:.3e}T + {c:.3e}T')
plt.xlabel('Temperatura (K)')
plt.ylabel('Cp/T (cal K)')
plt.title('Ajuste polinomial de 3 ordem')
plt.legend()
plt.grid(True)

# Grfico 2: Aproximao aT no melhor intervalo
plt.subplot(2, 2, 2)
mask_best = (T >= best_interval[0]) & (T <= best_interval[1])
T_best = T[mask_best]
Cp_T_best = Cp_T[mask_best]

T_plot_best = np.linspace(best_interval[0], best_interval[1], 100)
plt.plot(T_best, Cp_T_best, 'bo', label='Dados experimentais')
plt.plot(T_plot_best, cubic_law(T_plot_best, best_a), 'g-',
         label=f'Ajuste aT: a = {best_a:.3e}')
plt.xlabel('Temperatura (K)')
plt.ylabel('Cp/T (cal K)')
plt.title(f'Aproximao aT vlida em {best_interval[0]}-{best_interval[1]} K')
plt.legend()
plt.grid(True)

# Grfico 3: Comparao dos dois modelos
plt.subplot(2, 2, 3)
plt.plot(T, Cp_T, 'bo', label='Dados experimentais')
plt.plot(T_plot, fitted_function(T_plot), 'r-', label='Ajuste polinomial')
plt.plot(T_plot_best, cubic_law(T_plot_best, best_a), 'g--', label='Aproximao aT')
plt.xlabel('Temperatura (K)')
plt.ylabel('Cp/T (cal K)')
plt.title('Comparao dos modelos')
plt.legend()
plt.grid(True)

# Grfico 4: Resduos do ajuste polinomial
plt.subplot(2, 2, 4)
residuals = Cp_T - fitted_function(T)
plt.plot(T, residuals, 'ro')
plt.axhline(y=0, color='k', linestyle='--')
plt.xlabel('Temperatura (K)')
plt.ylabel('Resduos (Cp/T experimental - ajustado)')
plt.title('Resduos do ajuste polinomial')
plt.grid(True)

plt.tight_layout()
plt.show()

# Resultados finais
print("\n" + "="*60)
print("RESULTADOS FINAIS:")
print("="*60)
print(f"a) Entropia do NaSO a 90 K: {S_90K:.4f} cal K")
print(f"b) Aproximao aT  vlida no intervalo: {best_interval[0]}-{best_interval[1]} K")
print(f"   Parmetro a = {best_a:.6e}")
print(f"   Coeficiente de determinao R = {best_r2:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.optimize import curve_fit

# Dados fornecidos
T = np.array([13.74, 16.25, 20.43, 27.73, 41.11, 52.72, 68.15, 82.96, 95.71])  # K
Cp = np.array([0.171, 0.286, 0.626, 1.615, 4.346, 7.032, 10.48, 13.28, 15.33])  # cal/K

# Cp/T
Cp_T = Cp / T

# Ajuste polinomial de 3 ordem passando pela origem
def poly_3rd(x, a, b, c):
    return a*x + b*x**2 + c*x**3

params, _ = curve_fit(poly_3rd, T, Cp_T)
a, b, c = params

# Funo ajustada
def f(T):
    return poly_3rd(T, a, b, c)

# Entropia at 90 K
S_90K, _ = integrate.quad(f, 0, 90)

print("Parmetros do polinmio de 3 ordem:")
print(f"a = {a:.6e}, b = {b:.6e}, c = {c:.6e}")
print(f"Entropia do NaSO a 90 K = {S_90K:.4f} cal/K")

# Lei cbica Cp/T = aT em baixas T
def cubic(T, a):
    return a*T**3

params_cubic, _ = curve_fit(cubic, T[:3], Cp_T[:3])  # usando 3 primeiros pontos
a_cubic = params_cubic[0]

print(f"Aproximao Cp/T = aT em baixas T, a = {a_cubic:.6e}")

# Grfico Cp/T vs T com ajuste polinomial
T_plot = np.linspace(0, 100, 500)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(T, Cp_T, 'o', label='Dados experimentais')
plt.plot(T_plot, f(T_plot), '-', label='Ajuste polinomial')
plt.xlabel("T (K)")
plt.ylabel("Cp/T (cal K$^{-2}$)")
plt.title("Ajuste polinomial de 3 ordem")
plt.legend()
plt.grid(True)

# Grfico Cp/T vs T com aproximao cbica
plt.subplot(1,2,2)
plt.plot(T, Cp_T, 'o', label='Dados experimentais')
plt.plot(T, cubic(T, a_cubic), '--', label='Lei cbica')
plt.xlabel("T (K)")
plt.ylabel("Cp/T (cal K$^{-2}$)")
plt.title("Aproximao Cp/T = aT")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
